```python
import os
import subprocess
import collections
```

```python
stats = {"data_cache_const":{}, "ins_cache_const":{}}
```

```python
def get_data(file):
    data = {}
    with open(file, "r") as f:
        lines = f.readlines()
        data['no_of_instructions'] = int(lines[0].split("= ")[1].strip('\n'))
        data['no_of_cycles'] = int(lines[1].split("= ")[1].strip('\n'))
        data['throughput'] = float(lines[2].split("= ")[1].strip('\n'))
    return data
```

```python
for folder in os.listdir("stats"):
    if(folder[11:13] == 'id'):
        stats['ins_cache_const'][folder[13:]] = {}
        for file in os.listdir("stats/"+folder+"/"):
            if(file[-13:]== ".observedstat"):
                stats['ins_cache_const'][folder[13:]][file.split('.')[0]] = get_data("stats/"+folder+"/"+file)
    
    if(folder[11:13] == 'di'):
        stats['data_cache_const'][folder[13:]] = {}
        for file in os.listdir("stats/"+folder+"/"):
            if(file[-13:]== ".observedstat"):
                stats['data_cache_const'][folder[13:]][file.split('.')[0]] = get_data("stats/"+folder+"/"+file)
```

```python
import json

with open("stats.json", "w") as file_json:
    json.dump(stats, file_json)
```

```python
import matplotlib.pyplot as plt

data = {}
for program in stats['data_cache_const']['16']:
    ipcs = {}
    for key in stats['data_cache_const']:
        keyi = int(key)
        ipcs[keyi] = stats['data_cache_const'][key][program]['throughput']
    data[program] = collections.OrderedDict(sorted(ipcs.items()))
        
```

```python
for program in stats['data_cache_const']['16']:
    ipcs = data[program]
    plt.plot(ipcs.keys(), ipcs.values())
    plt.scatter(ipcs.keys(), ipcs.values(), c='red')
    plt.title(program)
    plt.savefig("figures/data_cache_const_"+ program +".png")
    plt.xlabel("Instruction cache size")
    plt.ylabel("throughput (IPC)")
    plt.show()
```
